```python
# Agregamos y definimos la función
def funcion(x):
    return np.cos(x) - 2*x + 3
```

```python
# Implementamos el Método de Steffensen
#Indicamos las variables
def Restricciones (func, p0, tol=0.00001, nmax=100):
    p = p0
    niter = 0
    error = float ('inf')
#Imprimimos el formato de tabla
    print("# iter\t\t p \t\t f(p) \t\t error")
    print("{0} \t\t {1:6.4f} \t {2:6.4f} \t {3:6.4f}".format(niter, p, func(p), error))
#Comenzamos con el bucle
    while error > tol and niter < nmax:
            fp = func(p)
            s  = p - fp
            fs = func(s)
            w  = p - (fp ** 2) / (fs - fp)
            error = abs(w  - p)
            p = w
            niter += 1
            print("{0} \t\t {1:6.6f} \t {2:6.6f} \t {3:6.6f}".format(niter, p, func(p), error))

    return p
```

```python
#Tomamos nuestro valor de referencia para la aproximacion
p0 = 1.0
```

```python
#concluimos el metodo
raiz = Restricciones (funcion, p0)
print("La raiz es {0:6.7f}:".format(raiz))
```
