```python
#Utilizamos la librería python
import sympy as sp
import matplotlib.pyplot as plt
import numpy as np
```

```python

# Función para calcular el polinomio de interpolación de Lagrange
def lagrange_interpolation(x_values, y_values):
    # Definir la variable simbólica x
    x = sp.symbols('x')

    # Inicializar el polinomio de interpolación
    n = len(x_values)
    P = 0

    # Calcular el polinomio de Lagrange
    for i in range(n):
        # Calcular el término de Lagrange Li(x)
        L = 1
        for j in range(n):
            if i != j:
                L *= (x - x_values[j]) / (x_values[i] - x_values[j])
        # Agregar el término al polinomio
        P += y_values[i] * L

    # Simplificar el polinomio
    P = sp.expand(P)

    return P, x

# Función para graficar el polinomio y los puntos de la tabla
def plot_lagrange(x_values, y_values, P, x):
    # Convertir el polinomio simbólico a función numérica
    P_func = sp.lambdify(x, P, 'numpy')

    # Crear valores de x para graficar el polinomio
    x_plot = np.linspace(min(x_values) - 1, max(x_values) + 1, 500)
    y_plot = P_func(x_plot)

    # Graficar el polinomio de interpolación
    plt.plot(x_plot, y_plot, label='Polinomio de Lagrange', color='pink')

    # Graficar los puntos de la tabla original
    plt.scatter(x_values, y_values, color='blue', label='Puntos de datos', zorder=5)

    # Añadir leyenda y títulos
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Interpolación de Lagrange')
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
# Ejemplo de uso
x_values = [4, 2, 1, 3]
y_values = [2, 4, 3, 1]

# Calcular el polinomio de interpolación de Lagrange
P, x = lagrange_interpolation(x_values, y_values)

# Mostrar el polinomio
print(f"Polinomio de Lagrange: {P}")

# Graficar el resultado
plot_lagrange(x_values, y_values, P, x)
```
